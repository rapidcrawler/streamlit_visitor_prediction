```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import shap
import matplotlib.pyplot as plt
```

```python
# Creating a dummy dataset
dummy_data = pd.DataFrame({
    'year': np.random.choice(range(2021, 2025), size=100),
    'month': np.random.choice(range(1, 13), size=100),
    'nationality': np.random.choice(['USA', 'UK', 'India', 'China', 'Germany'], size=100),
    'economy_ticket_price': np.random.uniform(200, 1500, size=100),
    'business_ticket_price': np.random.uniform(1000, 5000, size=100),
    'bookings_last_3_months': np.random.randint(50, 500, size=100),
    'bookings_last_2_months': np.random.randint(30, 400, size=100),
    'bookings_last_month': np.random.randint(10, 300, size=100)
})
```

```python
# Encode the categorical variables as the model expects
dummy_data['nationality_encoded'] = dummy_data['nationality'].astype('category').cat.codes
dummy_data
```

```python
# Prepare the input and output variables
X_dummy = dummy_data[['year', 'month', 'nationality_encoded', 'economy_ticket_price',
                      'business_ticket_price', 'bookings_last_3_months', 'bookings_last_2_months', 
                      'bookings_last_month']]
y_dummy = np.random.randint(1000, 5000, size=100)  # Random target variable for demonstration

# Train the model
model = LinearRegression()
model.fit(X_dummy, y_dummy)
```

```python
# Defining the values for Year, Month, and Nationality
years = [2024, 2025]
months = ["January", "February", "March", "April", "May", "June",
          "July", "August", "September", "October", "November", "December"]
nationalities = ["USA", "UK", "India", "China", "Germany"]

# Generating all possible combinations
combinations = list(itertools.product(years, months, nationalities))

# Creating a DataFrame with these combinations
inference_data = pd.DataFrame(combinations, columns=['year', 'month', 'nationality'])

# Adding random values for the rest of the columns
np.random.seed(0)  # For reproducibility
inference_data['economy_ticket_price'] = np.random.uniform(200, 1500, size=len(inference_data))
inference_data['business_ticket_price'] = np.random.uniform(1000, 5000, size=len(inference_data))
inference_data['bookings_last_3_months'] = np.random.randint(50, 500, size=len(inference_data))
inference_data['bookings_last_2_months'] = np.random.randint(30, 400, size=len(inference_data))
inference_data['bookings_last_month'] = np.random.randint(10, 300, size=len(inference_data))
inference_data
```

```python
inference_data_encoded = inference_data.copy()
inference_data_encoded['nationality_encoded'] = inference_data_encoded['nationality'].astype('category').cat.codes
# Define a dictionary to map month names to numbers
month_mapping = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12
}

# Apply the mapping to the month column
inference_data_encoded['month'] = inference_data_encoded['month'].map(month_mapping)
inference_data_encoded = inference_data_encoded[['year', 'month', 'nationality_encoded', 'economy_ticket_price',
                      'business_ticket_price', 'bookings_last_3_months', 'bookings_last_2_months', 
                      'bookings_last_month']]
# Display the updated dataframe
inference_data_encoded
```

```python
# Set Matplotlib backend to 'Agg' for non-interactive plotting
# Else exported SHAP plots will be blank white screen
matplotlib.use('Agg')

# Ensure the output directory exists
output_dir = './artifacts/shap_plots/'
os.makedirs(output_dir, exist_ok=True)

# Create a SHAP explainer using shap.Explainer
explainer = shap.Explainer(model, inference_data_encoded)

# Loop through each row in the inference dataset and plot SHAP waterfall plots
for i in range(len(inference_data_encoded)):
    # Calculate SHAP values for the current instance
    shap_values = explainer(inference_data_encoded.iloc[[i]])

    # Plot SHAP waterfall plot for the current instance
    shap.waterfall_plot(shap_values[0], max_display=12)

    # Save the plot to a file
    plot_path = os.path.join(output_dir, f"SHAP_Waterfall_{inference_data.iloc[i][0].astype(int)}_{inference_data.iloc[i][1].lower()}_{inference_data.iloc[i][2].lower()}.png")
    plt.savefig(plot_path, bbox_inches='tight')
    plt.close()

print(f"SHAP plots saved to: {output_dir}")
```
